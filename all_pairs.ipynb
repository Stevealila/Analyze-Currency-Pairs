```python
import MetaTrader5 as mt5
import pandas as pd
```

```python
mt5.initialize()
```

# All Symbols

```python
symbols = mt5.symbols_get()
df_symbols = pd.DataFrame(list(symbols), columns=symbols[0]._asdict().keys())

pd.set_option("display.max_row", 245)
pd.set_option("display.max_column", 97)

# Filter Forex symbols
forex_symbols = df_symbols[df_symbols['path'].str.contains('Forex')]
forex_symbols = forex_symbols[forex_symbols['name'].str.endswith('m')]
```

# Filtered Symbols

```python
target_columns = ['name', 'bid', 'bidhigh', 'bidlow', 'ask', 'askhigh', 'asklow', 'trade_tick_value']
df = forex_symbols[target_columns]
df = df.reset_index(drop=True)
df = df[df['bid'] != 0.00000]
df.head(2)
```

# Sort by highest trade tick value

```python
df = df.sort_values(by='trade_tick_value', ascending=False)
df = df.reset_index(drop=True)
df
```
