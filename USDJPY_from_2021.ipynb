```python
import MetaTrader5 as mt5
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pytz

mt5.initialize()
```

```python
timezone = pytz.timezone("Africa/Nairobi")
end_date = datetime.now(timezone)
# rates for the past 3 years
rates = mt5.copy_rates_from("USDJPYm", mt5.TIMEFRAME_D1, end_date, 1017)
print(rates[:5])
```

```python
pd.set_option('display.max_rows', 1001)
             
df = pd.DataFrame(rates)

df['time'] = pd.to_datetime(df['time'], unit='s')
df = df.drop(['real_volume', 'spread'], axis=1)
df.set_index('time', inplace=True)

df.shape
```

```python
df
```

```python
endmonth_reading = df.resample("ME").last()
endmonth_reading
```

```python
last_monthly_bz_day = df.resample("BME").last()
last_monthly_bz_day
```

```python
monthly_change = df.resample("BME").last().pct_change(periods=1).mul(100)
monthly_change = monthly_change.dropna()
monthly_change
```
